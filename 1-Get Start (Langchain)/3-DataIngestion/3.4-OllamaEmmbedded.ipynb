# OLLAMA EMBEDDING

```python
from langchain_community.embeddings import OllamaEmbeddings

embedding =(
    OllamaEmbeddings(model="deepseek-r1:8b") ## default llama
)
embedding
```

```python
r1 = embedding.embed_documents([
    "A is first letter of Greek alphabet",
    "Z is the last letter of Greek alphabet"
])

r1
```

```python
len(r1)
```

```python
r1 = embedding.embed_query("What is the last letter of english alphabet")
```
